```python
#Example adapted from https://github.com/rasbt/python-machine-learning-book/blob/master/code/ch12/ch12.ipynb
#Original Author: Sebastian Raschka
class TwoLayerPerceptronBase(object):
    def __init__(self, n_hidden=30, C=0.0, epochs=500, eta=0.001, random_state=None):
        np.random.seed(random_state)
        self.n_hidden = n_hidden
        self.l2_C = C
        self.epochs = epochs
        self.eta = eta
        
    @staticmethod
    def _encode_labels(y):
        """Encode labels into one-hot representation"""
        onehot = pd.get_dummies(y).values.T
            
        return onehot

    def _initialize_weights(self):
        """Initialize weights with small random numbers."""
        W1_num_elems = (self.n_features_)*self.n_hidden
        W1 = np.random.uniform(-1.0, 1.0, size=W1_num_elems)
        W1 = W1.reshape(self.n_hidden, self.n_features_) # reshape to be W
        b1 = np.zeros((self.n_hidden, 1))
        
        W2_num_elems = (self.n_hidden)*self.n_output_
        W2 = np.random.uniform(-1.0, 1.0, size=W2_num_elems)
        W2 = W2.reshape(self.n_output_, self.n_hidden)
        b2 = np.zeros((self.n_output_, 1))
        
        return W1, W2, b1, b2
    
    @staticmethod
    def _sigmoid(z):
        """Use scipy.special.expit to avoid overflow"""
        # 1.0 / (1.0 + np.exp(-z))
        return expit(z)
    
    
    @staticmethod
    def _L2_reg(lambda_, W1, W2):
        """Compute L2-regularization cost"""
        # only compute for non-bias terms
        return (lambda_/2.0) * np.sqrt(np.mean(W1[:, 1:] ** 2) + np.mean(W2[:, 1:] ** 2))
    
    def _cost(self,A3,Y_enc,W1,W2):
        '''Get the objective function value'''
        cost = np.mean((Y_enc-A3)**2)
        L2_term = self._L2_reg(self.l2_C, W1, W2)
        return cost + L2_term
    
    def _feedforward(self, X, W1, W2, b1, b2):
        """Compute feedforward step
        -----------
        X : Input layer with original features.
        W1: Weight matrix for input layer -> hidden layer.
        W2: Weight matrix for hidden layer -> output layer.
        ----------
        a1-a3 : activations into layer (or output layer)
        z1-z2 : layer inputs 

        """
        A1 = X.T
        Z1 = W1 @ A1 + b1
        A2 = self._sigmoid(Z1)
        Z2 = W2 @ A2 + b2
        A3 = self._sigmoid(Z2)
        return A1, Z1, A2, Z2, A3
    
    def _get_gradient(self, A1, A2, A3, Z1, Z2, Y_enc, W1, W2):
        """ Compute gradient step using backpropagation.
        """
        # vectorized backpropagation
        V2 = -2*(Y_enc-A3)*A3*(1-A3)
        V1 = A2*(1-A2)*(W2.T @ V2)
        
        gradW2 = V2 @ A2.T
        gradW1 = V1 @ A1.T
        
        gradb2 = np.sum(V2, axis=1).reshape((-1,1))
        gradb1 = np.sum(V1, axis=1).reshape((-1,1))
        
        # regularize weights that are not bias terms
        gradW1 += W1 * self.l2_C
        gradW2 += W2 * self.l2_C

        return gradW1, gradW2, gradb1, gradb2
    
    def predict(self, X):
        """Predict class labels"""
        _, _, _, _, A3 = self._feedforward(X, self.W1, self.W2, self.b1, self.b2)
        y_pred = np.argmax(A3, axis=0)
        return y_pred
```

```python
class TLPMiniBatch(TwoLayerPerceptronBase):
    def __init__(self, alpha=0.0, decrease_const=0.0, shuffle=True, minibatches=1, **kwds):

        # need to add to the original initializer 
        self.alpha = alpha
        self.decrease_const = decrease_const
        self.shuffle = shuffle
        self.minibatches = minibatches
        
        # but keep other keywords
        super().__init__(**kwds)
        
    
    def fit(self, X, y, print_progress=False, XY_test=None):
        """ Learn weights from training data. With mini-batch"""
        X_data, y_data = X.copy(), y.copy()
        Y_enc = self._encode_labels(y)
        
        # init weights and setup matrices
        self.n_features_ = X_data.shape[1]
        self.n_output_ = Y_enc.shape[0]
        self.W1, self.W2, self.b1, self.b2 = self._initialize_weights()

        # start momentum at zero for previous updates
        rho_W1_prev = np.zeros(self.W1.shape)
        rho_W2_prev = np.zeros(self.W2.shape)
        rho_b1_prev = np.zeros(self.b1.shape)
        rho_b2_prev = np.zeros(self.b2.shape)

        self.cost_ = []
        self.score_ = []
        # get starting acc
        self.score_.append(accuracy_score(y_data,self.predict(X_data)))
        # keep track of validation, if given
        # provide tuple of this
        # this tracks the health of validation
        if XY_test is not None:
            X_test = XY_test[0].copy()
            y_test = XY_test[1].copy()
            self.val_score_ = []
            self.val_score_.append(accuracy_score(y_test,self.predict(X_test)))
            
        for i in range(self.epochs):

            # adaptive learning rate
            eta = self.eta / (1 + self.decrease_const*i)

            if print_progress>0 and (i+1)%print_progress==0:
                sys.stderr.write('\rEpoch: %d/%d' % (i+1, self.epochs))
                sys.stderr.flush()

            if self.shuffle:
                idx_shuffle = np.random.permutation(y_data.shape[0])
                X_data, Y_enc, y_data = X_data[idx_shuffle], Y_enc[:, idx_shuffle], y_data[idx_shuffle]

            mini = np.array_split(range(y_data.shape[0]), self.minibatches)
            mini_cost = []
            for idx in mini:

                # feedforward
                A1, Z1, A2, Z2, A3 = self._feedforward(X_data[idx],
                                                       self.W1,
                                                       self.W2,
                                                       self.b1,
                                                       self.b2
                                                      )
                
                cost = self._cost(A3,Y_enc[:, idx],self.W1,self.W2)
                mini_cost.append(cost) # this appends cost of mini-batch only

                # compute gradient via backpropagation
                gradW1, gradW2, gradb1, gradb2 = self._get_gradient(A1=A1, A2=A2, A3=A3, Z1=Z1, Z2=Z2, 
                                                  Y_enc=Y_enc[:, idx],
                                                  W1=self.W1,W2=self.W2)
                
                # momentum calculations
                rho_W1, rho_W2 = eta * gradW1, eta * gradW2
                rho_b1, rho_b2 = eta * gradb1, eta * gradb2
                self.W1 -= (rho_W1 + (self.alpha * rho_W1_prev)) # update with momentum
                self.W2 -= (rho_W2 + (self.alpha * rho_W2_prev)) # update with momentum
                self.b1 -= (rho_b1 + (self.alpha * rho_b1_prev))
                self.b2 -= (rho_b2 + (self.alpha * rho_b2_prev))
                rho_W1_prev, rho_W2_prev = rho_W1, rho_W2
                rho_b1_prev, rho_b2_prev = rho_b1, rho_b2


            self.cost_.append(mini_cost)
            self.score_.append(accuracy_score(y_data,self.predict(X_data)))
            
            # calculate the validation accuracy
            if XY_test is not None:
                self.val_score_.append(accuracy_score(y_test,self.predict(X_test)))
            
        return self
```

```python
class TLPMiniBatchCrossEntropy(TLPMiniBatch):
    def _cost(self,A3,Y_enc,W1,W2):
        '''Get the objective function value'''
        cost = -np.mean(np.nan_to_num((Y_enc*np.log(A3)+(1-Y_enc)*np.log(1-A3))))
        L2_term = self._L2_reg(self.l2_C, W1, W2)
        return cost + L2_term
    
    def _get_gradient(self, A1, A2, A3, Z1, Z2, Y_enc, W1, W2):
        """ Compute gradient step using backpropagation.
        """
        # vectorized backpropagation
        # V2 represents the cross entropy calculation (negative log likelihood)
        V2 = (A3-Y_enc) # <- this is only line that changed
        V1 = A2*(1-A2)*(W2.T @ V2)
        
        gradW2 = V2 @ A2.T
        gradW1 = V1 @ A1.T
        
        gradb2 = np.sum(V2, axis=1).reshape((-1,1))
        gradb1 = np.sum(V1, axis=1).reshape((-1,1))
        
        # regularize weights that are not bias terms
        gradW1 += W1 * self.l2_C
        gradW2 += W2 * self.l2_C

        return gradW1, gradW2, gradb1, gradb2
```

```python
class TLPBetterInitial(TLPMiniBatchCrossEntropy):             
    def _initialize_weights(self):
        """Initialize weights Glorot and He normalization."""
        init_bound = 4*np.sqrt(6. / (self.n_hidden + self.n_features_))
        W1 = np.random.uniform(-init_bound, init_bound,(self.n_hidden, self.n_features_))

        # reduce the final layer magnitude in order to balance the size of the gradients
        # between 
        init_bound = 4*np.sqrt(6 / (self.n_output_ + self.n_hidden))
        W2 = np.random.uniform(-init_bound, init_bound,(self.n_output_, self.n_hidden)) 
        
        b1 = np.zeros((self.n_hidden, 1))
        b2 = np.zeros((self.n_output_, 1))
        
        return W1, W2, b1, b2
```

```python
# print out accuracy results
def print_result(nn,X_train,y_train,X_test,y_test,title="",color="red"):
    print("=================")
    print(title,":")
    yhat = nn.predict(X_train)
    print('Resubstitution acc:',accuracy_score(y_train,yhat))
    
    yhat = nn.predict(X_test)
    accuracy = accuracy_score(y_test,yhat)
    print('Validation acc:',accuracy)
    
    print("Precision Score:", precision_score(y_test, yhat, average="weighted"))
    print("Recall Score:", recall_score(y_test, yhat, average="weighted"))
    print("F1 Score:", f1_score(y_test, yhat, average="weighted"))
    print("Model error:", 1 - accuracy)
    
    # calculate confusion matrix and plot
    cm = confusion_matrix(y_test, yhat)
    cm_disp = ConfusionMatrixDisplay(cm, display_labels=[0,1,2,3])
    cm_disp.plot()
    plt.grid(False)
    plt.show()
    
    if hasattr(nn,'val_score_'):
        plt.plot(range(len(nn.val_score_)), nn.val_score_, color=color,label=title)
        plt.ylabel('Validation Accuracy')
    else:
        plt.plot(range(len(nn.score_)), nn.score_, color=color,label=title)
        plt.ylabel('Resub Accuracy')
        
    plt.xlabel('Epochs')
    plt.tight_layout()
    plt.legend(loc='best')
    plt.grid(True)
```

Instead of manually creating a specific class for a 3 layer network, I implemented for loops to iterate through each layer to update the weights and biases.  A, Z, W, b, and the resepctive grads of W and b will all be held in a matrix - each will be a matrix of matrices where each inner matrix represents the values at that layer (mostly).  Later variables like V (sensitivity, not AdaM) have it reversed where the first index is the last layer in the V matrix, as opposed to, for example, the W matrix which the first index corresponds to the first layer and so on.

```python
class MultiLayerPerceptron():
    def __init__(self, n_layers=3, n_hidden=[50,30], C=0.0, epochs=500, eta=0.001, random_state=None):
        np.random.seed(random_state)
        self.n_layers = n_layers  # new parameter for dynamic neural networks
        self.n_hidden = n_hidden
        self.l2_C = C
        self.epochs = epochs
        self.eta = eta
        
    @staticmethod
    def _encode_labels(y):
        """Encode labels into one-hot representation"""
        onehot = pd.get_dummies(y).values.T
            
        return onehot

    def _initialize_weights(self):
        """
        Initialize weights with small random numbers.
        Changes made:
            Created empty list to store W and b
            Idea is to be dynamic via n_layers by storing all initialized W and b 
            as a list in each list
            Last W must be configured with the last output number of neurons, 
            hence not part of the loop
        """
        # initialize an empty list to store all initialized weights
        All_W = []
        
        # initialize an empty list to store all initialized biases (should all be 1)
        All_b = []
        
        # initialize first layer W
        W_num_elems = (self.n_features_)*self.n_hidden[0]
        W_first = np.random.uniform(-1.0, 1.0, size=W_num_elems)
        W_first = W_first.reshape(self.n_hidden[0], self.n_features_) # reshape to be W
        
        All_W.append(W_first)
        
        # initialize first layer b
        b_first = np.zeros((self.n_hidden[0], 1))
        
        All_b.append(b_first)
        
        # iterate through layers
        for layer in range(self.n_layers - 2):  # -2 because we already have 1st and last layer
            W_num_elems = (self.n_hidden[layer])*self.n_hidden[layer + 1]
            W = np.random.uniform(-1.0, 1.0, size=W_num_elems)
            W = W.reshape(self.n_hidden[layer + 1], self.n_hidden[layer]) # reshape to be W
            b = np.zeros((self.n_hidden[layer + 1], 1))
            
            # append W and b
            All_W.append(W)
            All_b.append(b)
        
        # intialize last layer
        W_last_num_elems = (self.n_hidden[-1])*self.n_output_
        W_last = np.random.uniform(-1.0, 1.0, size=W_last_num_elems)
        W_last = W_last.reshape(self.n_output_, self.n_hidden[-1])
        b_last = np.zeros((self.n_output_, 1))
        
        All_W.append(W_last)
        All_b.append(b_last)
        
        return All_W, All_b
    
    @staticmethod
    def _sigmoid(z):
        """Use scipy.special.expit to avoid overflow"""
        # 1.0 / (1.0 + np.exp(-z))
        return expit(z)
    
    
    @staticmethod
    def _L2_reg(lambda_, W):
        """
        Compute L2-regularization cost
        Changes made:
            For L2, we now need to take the summation of all 
            W now that we have a dynamic network
        """
        
        # iterate through each weight and find sum square
        total_weight = 0
        for weight in W:
            total_weight += np.mean(weight[:, 1:] ** 2)
        
        # only compute for non-bias terms
        return (lambda_/2.0) * np.sqrt(total_weight)
    
    def _cost(self, A_last, Y_enc, W):
        '''
        Get the objective function value
        Changes made:
            Need to pass through W as W can be dynamic 
            depending on size of network
            Renamed A3 to A_last for consistency
        '''
        cost = np.mean((Y_enc-A_last)**2)
        L2_term = self._L2_reg(self.l2_C, W)
        return cost + L2_term
    
    def _feedforward(self, X, W, b):
        """Compute feedforward step
        -----------
        X : Input layer with original features.
        
        W: Weight matrix of matrix for all layers
            1st matrix = first layer -> hidden layer1
            2nd matrix = hidden layer1 -> hidden layer2
            ...etc
            
        b: bias matrix of matrix for all layers
            1st matrix = first layer -> hidden layer1
            2nd matrix = hidden layer1 -> hidden layer2
            ...etc
        ----------
        A: Activations matrix by each layer
        Z: Layer inputs matrix by each layer

        """
        # initialize list of A and Z to store each layer A and Z
        A = []
        Z = []
        
        # first layer A - store it
        A_first = X.T
        A.append(A_first)
        
        # iterate through layers
        for layer in range(self.n_layers - 1):  # once again, need to skip the last layer
            # calculate Z
            Z_layer = W[layer] @ A[layer] + b[layer]
            Z.append(Z_layer)
            
            # calculate A
            A_layer = self._sigmoid(Z_layer)
            A.append(A_layer)
        
        
        # last layer Z
        Z_last = W[self.n_layers - 1] @ A[self.n_layers - 1] + b[self.n_layers - 1]
        Z.append(Z_last)
        
        # last layer A
        A_last = self._sigmoid(Z_last)
        A.append(A_last)
        
        return A, Z
    
    def _get_gradient(self, A, Z, Y_enc, W):
        """
        Compute gradient step using backpropagation
        Changes made:
            Added V, which is the master list of all V to store
            Added grad_w, which is the master list of all grad for w to store
            Added grad_b, which is the master list of all grade for b to store
        """
        # initialize master lists
        V = []
        grad_W = []
        grad_b = []
        
        # calculate the last layer V, W, b
        V_last = -2*(Y_enc-A[-1])*A[-1]*(1-A[-1])
        gradW_last = V_last @ A[-2].T
        gradb_last = np.sum(V_last, axis=1).reshape((-1,1))
        
        # append last V, W, b
        V.append(V_last)
        grad_W.append(gradW_last)
        grad_b.append(gradb_last)
        
        # iterate through layers
        # need to reverse because it back propogates
        # also need to go from last layer to 1st layer (0 isn't a layer)
        for layer in reversed(range(1, self.n_layers)):
            V_idx = abs(layer - self.n_layers + 1)

            # update V
            V_layer = A[layer] * (1 - A[layer]) * (W[layer].T @ V[V_idx])
            V.append(V_layer)
            
            # update W
            gradW_layer = V[V_idx + 1] @ A[layer-1].T
            
            # regularize weights that are not bias terms
            gradW_layer += W[layer-1] * self.l2_C
            grad_W.insert(0, gradW_layer)
            
            # update b
            gradb_layer = np.sum(V[V_idx + 1], axis=1).reshape((-1,1))
            grad_b.insert(0, gradb_layer)

        return grad_W, grad_b
    
    def predict(self, X):
        """Predict class labels"""
        A, Z = self._feedforward(X, self.W, self.b)
        y_pred = np.argmax(A[self.n_layers], axis=0)
        return y_pred
```

```python
class MLPMiniBatch(MultiLayerPerceptron):
    def __init__(self, alpha=0.0, decrease_const=0.0, shuffle=True, minibatches=1, **kwds):

        # need to add to the original initializer 
        self.alpha = alpha
        self.decrease_const = decrease_const
        self.shuffle = shuffle
        self.minibatches = minibatches
        
        # but keep other keywords
        super().__init__(**kwds)
        
    
    def fit(self, X, y, print_progress=False, XY_test=None):
        """ Learn weights from training data. With mini-batch"""
        X_data, y_data = X.copy(), y.copy()
        Y_enc = self._encode_labels(y)
        
        # init weights and setup matrices
        self.n_features_ = X_data.shape[1]
        self.n_output_ = Y_enc.shape[0]
        self.W, self.b = self._initialize_weights()

        # start momentum at zero for previous updates
        rho_W = []
        rho_W_prev = []
        rho_b = []
        rho_b_prev = []
        
        # iterate through layers and initialize momentum
        for layer in range(self.n_layers):
            rho_W.append(np.zeros(self.W[layer].shape))
            rho_W_prev.append(np.zeros(self.W[layer].shape))
            rho_b.append(np.zeros(self.b[layer].shape))
            rho_b_prev.append(np.zeros(self.b[layer].shape))

        self.cost_ = []
        self.score_ = []
        # get starting acc
        self.score_.append(accuracy_score(y_data,self.predict(X_data)))
        # keep track of validation, if given
        # provide tuple of this
        # this tracks the health of validation
        if XY_test is not None:
            X_test = XY_test[0].copy()
            y_test = XY_test[1].copy()
            self.val_score_ = []
            self.val_score_.append(accuracy_score(y_test,self.predict(X_test)))
            
        for i in range(self.epochs):

            # adaptive learning rate
            eta = self.eta / (1 + self.decrease_const*i)

            if print_progress>0 and (i+1)%print_progress==0:
                sys.stderr.write('\rEpoch: %d/%d' % (i+1, self.epochs))
                sys.stderr.flush()

            if self.shuffle:
                idx_shuffle = np.random.permutation(y_data.shape[0])
                X_data, Y_enc, y_data = X_data[idx_shuffle], Y_enc[:, idx_shuffle], y_data[idx_shuffle]

            mini = np.array_split(range(y_data.shape[0]), self.minibatches)
            mini_cost = []
            for idx in mini:

                # feedforward - changed here because function changed in inherited class
                A, Z = self._feedforward(X_data[idx], self.W, self.b)
                
                # calculate cose - changed here because function changed above
                cost = self._cost(A[-1],Y_enc[:, idx], self.W)
                mini_cost.append(cost) # this appends cost of mini-batch only

                # compute gradient via backpropagation
                grad_W, grad_b = self._get_gradient(A=A, Z=Z, Y_enc=Y_enc[:, idx], W=self.W)
                
                # iterate through the network
                for layer in range(self.n_layers):

                    # momentum calculations
                    rho_W[layer] = eta * grad_W[layer]
                    rho_b[layer] = eta * grad_b[layer]      
                    self.W[layer] -= (rho_W[layer] + (self.alpha * rho_W_prev[layer])) # update with momentum
                    self.b[layer] -= (rho_b[layer] + (self.alpha * rho_b_prev[layer]))
                    rho_W_prev[layer] = rho_W[layer]
                    rho_b_prev[layer] = rho_b[layer]

            self.cost_.append(mini_cost)
            self.score_.append(accuracy_score(y_data,self.predict(X_data)))
            
            # calculate the validation accuracy
            if XY_test is not None:
                self.val_score_.append(accuracy_score(y_test,self.predict(X_test)))
            
        return self
```

```python
class MLPMiniBatchCrossEntropy(MLPMiniBatch):
    def _cost(self, A, Y_enc, W):
        '''
        Get the objective function value
        Changes made:
            Made dynamic the A parameter to point to last layer
            Corrected l2 parameter signature to take in W instead of individual Ws
        '''
        cost = -np.mean(np.nan_to_num((Y_enc*np.log(A[-1])+(1-Y_enc)*np.log(1-A[-1]))))
        L2_term = self._L2_reg(self.l2_C, W)
        return cost + L2_term
    
    def _cost(self, A_last, Y_enc, W):
        '''
        Get the objective function value
        Changes made:
            Need to pass through W as W can be dynamic 
            depending on size of network
            Renamed A3 to A_last for consistency
        '''
        cost = np.mean((Y_enc-A_last)**2)
        L2_term = self._L2_reg(self.l2_C, W)
        return cost + L2_term
    
    def _get_gradient(self, A, Z, Y_enc, W):
        """ 
        Compute gradient step using backpropagation.
        Changes made:
            Changed V_last to be A[last layer] - Y_enc
        """
        # initialize master lists
        V = []
        grad_W = []
        grad_b = []
        
        # calculate the last layer V, W, b
        V_last = (A[-1] - Y_enc)
        gradW_last = V_last @ A[-2].T
        gradb_last = np.sum(V_last, axis=1).reshape((-1,1))
        
        # append last V, W, b
        V.append(V_last)
        grad_W.append(gradW_last)
        grad_b.append(gradb_last)
        
        # iterate through layers
        # need to reverse because it back propogates
        # also need to go from last layer to 1st layer (0 isn't a layer)
        for layer in reversed(range(1, self.n_layers)):
            V_idx = abs(layer - self.n_layers + 1)

            # update V
            V_layer = A[layer] * (1 - A[layer]) * (W[layer].T @ V[V_idx])
            V.append(V_layer)
            
            # update W
            gradW_layer = V[V_idx + 1] @ A[layer-1].T
            
            # regularize weights that are not bias terms
            gradW_layer += W[layer-1] * self.l2_C
            grad_W.insert(0, gradW_layer)
            
            # update b
            gradb_layer = np.sum(V[V_idx + 1], axis=1).reshape((-1,1))
            grad_b.insert(0, gradb_layer)

        return grad_W, grad_b
```

```python
class MLPBetterInitial(MLPMiniBatchCrossEntropy):             
    def _initialize_weights(self):
        """Initialize weights Glorot and He normalization."""
        # initialize empty list to store all weights
        All_W = []
        
        # initialize empty list to store all bias
        All_b = []
        
        # initialize first layer weights with Glorot and He normalization
        init_bound = 4*np.sqrt(6. / (self.n_hidden[0] + self.n_features_))
        W_first = np.random.uniform(-init_bound, init_bound,(self.n_hidden[0], self.n_features_))

        # append to master W
        All_W.append(W_first)
        
        # initialize first layer bias
        b_first = np.zeros((self.n_hidden[0], 1))
        
        # append to master b
        All_b.append(b_first)
        
        # iterate through the layers
        for layer in range(self.n_layers - 2):  # Subtract 2 because we are skipping 1st and last layer
            
            # initialize hidden layer weights
            init_bound = 4*np.sqrt(6. / (self.n_hidden[layer + 1] + self.n_hidden[layer]))
            W_mid = np.random.uniform(-init_bound, init_bound,(self.n_hidden[layer + 1], self.n_hidden[layer]))
            All_W.append(W_mid)
            
            # initialize hidden layer bias
            b_mid = np.zeros((self.n_hidden[layer + 1], 1))
            All_b.append(b_mid)
        
        # reduce the final layer magnitude in order to balance the size of the gradients
        # between 
        last_bound = 4*np.sqrt(6 / (self.n_output_ + self.n_hidden[-1]))
        W_last = np.random.uniform(-last_bound, last_bound,(self.n_output_, self.n_hidden[-1]))
        
        # append to master W
        All_W.append(W_last)
        
        # finalize bias
        b_last = np.zeros((self.n_output_, 1))
        All_b.append(b_last)
        
        return All_W, All_b
```

```python
class MLPAdaptiveLearning(MLPBetterInitial):
    def __init__(self, beta1=0.9, beta2=0.999, epsilon=1e-8, optimizer="RMSProp", **kwds):

        # need to add to the original initializer 
        self.beta1 = beta1
        self.beta2 = beta2
        self.epsilon = epsilon
        self.optimizer = optimizer
        
        # but keep other keywords
        super().__init__(**kwds)
        
    
    def fit(self, X, y, print_progress=False, XY_test=None):
        """ Learn weights from training data. With mini-batch"""
        X_data, y_data = X.copy(), y.copy()
        Y_enc = self._encode_labels(y)
        
        # init weights and setup matrices
        self.n_features_ = X_data.shape[1]
        self.n_output_ = Y_enc.shape[0]
        self.W, self.b = self._initialize_weights()

        # start momentum at zero for previous updates
        rho_W = []
        rho_W_prev = []
        rho_b = []
        rho_b_prev = []
        g = []
        g_prev = []
        v = []
        v_prev = []
        m = []
        m_prev = []
        m_hat = []
        v_hat = []
        
        # iterate through layers and initialize momentum
        for layer in range(self.n_layers):
            rho_W.append(np.zeros(self.W[layer].shape))
            rho_W_prev.append(np.zeros(self.W[layer].shape))
            rho_b.append(np.zeros(self.b[layer].shape))
            rho_b_prev.append(np.zeros(self.b[layer].shape))
            g.append(np.zeros(self.W[layer].shape))
            g_prev.append(np.zeros(self.W[layer].shape))
            v.append(np.zeros(self.W[layer].shape))
            v_prev.append(np.zeros(self.W[layer].shape))
            m.append(np.zeros(self.W[layer].shape))
            m_prev.append(np.zeros(self.W[layer].shape))
            m_hat.append(np.zeros(self.W[layer].shape))
            v_hat.append(np.zeros(self.W[layer].shape))

        self.cost_ = []
        self.score_ = []
        # get starting acc
        self.score_.append(accuracy_score(y_data,self.predict(X_data)))
        # keep track of validation, if given
        # provide tuple of this
        # this tracks the health of validation
        if XY_test is not None:
            X_test = XY_test[0].copy()
            y_test = XY_test[1].copy()
            self.val_score_ = []
            self.val_score_.append(accuracy_score(y_test,self.predict(X_test)))
            
        for i in range(self.epochs):

            # adaptive learning rate
            eta = self.eta / (1 + self.decrease_const*i)

            if print_progress>0 and (i+1)%print_progress==0:
                sys.stderr.write('\rEpoch: %d/%d' % (i+1, self.epochs))
                sys.stderr.flush()

            if self.shuffle:
                idx_shuffle = np.random.permutation(y_data.shape[0])
                X_data, Y_enc, y_data = X_data[idx_shuffle], Y_enc[:, idx_shuffle], y_data[idx_shuffle]

            mini = np.array_split(range(y_data.shape[0]), self.minibatches)
            mini_cost = []
            for j, idx in enumerate(mini):

                # feedforward - changed here because function changed in inherited class
                A, Z = self._feedforward(X_data[idx], self.W, self.b)
                
                # calculate cose - changed here because function changed above
                cost = self._cost(A[-1],Y_enc[:, idx], self.W)
                mini_cost.append(cost) # this appends cost of mini-batch only

                # compute gradient via backpropagation
                grad_W, grad_b = self._get_gradient(A=A, Z=Z, Y_enc=Y_enc[:, idx], W=self.W)
                
                # iterate through the network
                for layer in range(self.n_layers):

                    # update W based on optimizer
                    if self.optimizer == "AdaGrad":
                        
                        # momentum calculations AdaGrad
                        g[layer] = self.beta1 * g_prev[layer] + (grad_W[layer] ** 2) 
                        rho_W[layer] = 1/np.sqrt(g[layer] + self.epsilon) * grad_W[layer]
                        
                    if self.optimizer == "RMSProp":
                        
                        # momentum calculations RMSProp
                        g[layer] = (grad_W[layer] ** 2)
                        v[layer] = self.beta1 * v_prev[layer] + (1 - self.beta1) * g[layer]
                        rho_W[layer] = 1/np.sqrt(v[layer] + self.epsilon) * grad_W[layer]
                    
                    if self.optimizer == "AdaDelta":
                        
                        # momentum calculations AdaDelta
                        g[layer] = (grad_W[layer] ** 2)
                        v[layer] = self.beta1 * v_prev[layer] + (1 - self.beta1) * g[layer]
                        m[layer] = self.beta1 * m_prev[layer] + (1 - self.beta1) * grad_W[layer]
                        rho_W[layer] = 1/np.sqrt(v[layer] + self.epsilon) * m[layer]
                    
                    if self.optimizer == "AdaM":
                        # momentum calculations AdaM
                        g[layer] = (grad_W[layer] ** 2)
                        m[layer] = self.beta1 * m_prev[layer] + (1 - self.beta1) * grad_W[layer]
                        v[layer] = self.beta2 * v_prev[layer] + (1 - self.beta2) * g[layer]
                        m_hat[layer] = m[layer]/(1 - self.beta1**(layer+1))
                        v_hat[layer] = v[layer]/(1 - self.beta2**(layer+1))
                        rho_W[layer] = m_hat[layer]/np.sqrt(v_hat[layer] + self.epsilon)

                    # update bias momentum
                    rho_b[layer] = eta * grad_b[layer]
                    
                    # update weights and bias
                    self.W[layer] -= (eta * rho_W[layer])
                    self.b[layer] -= (rho_b[layer] + (self.alpha * rho_b_prev[layer]))
                    rho_W_prev[layer] = rho_W[layer]
                    rho_b_prev[layer] = rho_b[layer]
                    g_prev[layer] = g[layer]
                    v_prev[layer] = v[layer]
                    m_prev[layer] = m[layer]

            self.cost_.append(mini_cost)
            self.score_.append(accuracy_score(y_data,self.predict(X_data)))
            
            # calculate the validation accuracy
            if XY_test is not None:
                self.val_score_.append(accuracy_score(y_test,self.predict(X_test)))
            
        return self
```
